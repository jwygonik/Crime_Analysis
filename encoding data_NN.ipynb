```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
ml_df = pd.read_csv("Resources/la_arrests_2020_w_zip.csv")
ml_df.head(20)
```

```python
ml_df = ml_df.drop(columns=['Booking Location','Location'])
ml_df.head(20)
```

```python
ml_df_two = ml_df.drop(columns=['Charge Description'])
```

```python
def agegroups(x):
    if x < 18:
        return 'Under_18'
    elif x <= 25:
        return '18-25'
    elif x <= 35:
        return '26-35'
    elif x <= 45:
        return '36-45'
    elif x <= 55:
        return '46-55'
    elif x <= 65:
        return '56-65'
    else:
        return '65_Plus'

ml_df_two['Age_Groups'] = ml_df_two['Age'].apply(agegroups)
ml_df_two.head()
```

```python
ml_df_two = ml_df_two.drop(columns=['Age','Area Name'])
ml_df_two.head()
```

```python
ml_cat_1 = ml_df_two.dtypes[ml_df_two.dtypes=='object'].index.tolist()
ml_cat_2 = ml_df_two.dtypes[ml_df_two.dtypes=='int64'].index.tolist()
ml_cat_3 = ml_df_two.dtypes[ml_df_two.dtypes=='float64'].index.tolist()
ml_cat = ml_cat_1 + ml_cat_2 + ml_cat_3
ml_df_two[ml_cat].nunique()
```

```python
import datetime as dt 
ml_df_two['Arrest Date'] = pd.to_datetime(ml_df_two['Arrest Date'])
ml_df_two.dtypes
```

```python
ml_df_two['Month_of_Arrest'] = pd.DatetimeIndex(ml_df_two['Arrest Date']).month
ml_df_two.head()
```

```python
ml_df_two = ml_df_two.drop(columns=['Arrest Date','LAT','LON'])
ml_df_two.head()
```

```python
ml_cat_1 = ml_df_two.dtypes[ml_df_two.dtypes=='object'].index.tolist()
ml_cat_2 = ml_df_two.dtypes[ml_df_two.dtypes=='int64'].index.tolist()
ml_cat_3 = ml_df_two.dtypes[ml_df_two.dtypes=='float64'].index.tolist()
ml_cat = ml_cat_1 + ml_cat_2 + ml_cat_3
ml_df_two[ml_cat].nunique()
```

```python
ml_df_two = ml_df_two.rename(columns={'Charge Group Description':'Charge_Group_Description','Disposition Description':'Disposition_Description','Descent Code':'Descent_Code'})
```

```python
# Determine which values to replace if counts are less than ...?
replace = list(binning[binning<1000].index)

# Replace in dataframe
for item in replace:
    ml_df_two.Charge_Group_Description = ml_df_two.Charge_Group_Description.replace(item,"Other")
    
# Check to make sure binning was successful
ml_df_two.Charge_Group_Description.value_counts()
```

```python
# Determine which values to replace if counts are less than ...?
replace = list(binning[binning<2000].index)

# Replace in dataframe
for item in replace:
    ml_df_two.Descent_Code = ml_df_two.Descent_Code.replace(item,"Other")
    
# Check to make sure binning was successful
ml_df_two.Descent_Code.value_counts()
```

```python
binning = ml_df_two.Disposition_Description.value_counts()
binning
```

```python
# Determine which values to replace if counts are less than ...?
replace = list(binning[binning<1000].index)

# Replace in dataframe
for item in replace:
    ml_df_two.Disposition_Description = ml_df_two.Disposition_Description.replace(item,"Other")
    
# Check to make sure binning was successful
ml_df_two.Disposition_Description.value_counts()
```

```python
ml_cat_1 = ml_df_two.dtypes[ml_df_two.dtypes=='object'].index.tolist()
ml_cat_2 = ml_df_two.dtypes[ml_df_two.dtypes=='int64'].index.tolist()
ml_cat_3 = ml_df_two.dtypes[ml_df_two.dtypes=='float64'].index.tolist()
ml_cat = ml_cat_1 + ml_cat_2 + ml_cat_3
ml_df_two[ml_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(ml_df_two[ml_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(ml_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
new_ml_df = ml_df_two.merge(encode_df,left_index=True, right_index=True)
new_ml_df = new_ml_df.drop(ml_cat,1)
new_ml_df.head()
```

```python
new_ml_df = new_ml_df.drop(columns=['Sex Code_F'])
new_ml_df = new_ml_df.rename(columns={'Sex Code_M':'Sex_Code_M'})

new_ml_df.to_csv('encoded_data.csv', index = False)
```

```python
y = new_ml_df.Sex_Code_M
X = new_ml_df.drop(columns=['Sex_Code_M'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 100
hidden_nodes_layer2 = 100
hidden_nodes_layer3 = 100

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="tanh"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the checkpoint path and filenames
os.makedirs("checkpoints/",exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}.hdf5"
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
batch_size=1135
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=0,
    save_weights_only=True,
    save_freq=5*batch_size)
```

```python
# Train the model

fit_model = nn.fit(X_train_scaled, y_train, epochs=50,callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
