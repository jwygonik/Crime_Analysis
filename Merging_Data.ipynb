```python
# Import dependencies
import pandas as pd
```

```python
# Import Arrest CSV
arrest_file = pd.read_csv('Cleaned_Data/la_arrests_2020_w_zip.csv')
arrest_df = pd.DataFrame(arrest_file)
arrest_df
```

```python
pd_file = pd.read_csv('Cleaned_Data/LAPD_Stats.csv')
pd_df = pd.DataFrame(pd_file)
pd_df
```

```python
# Create new DataFrame w/ police and arrest csv
police_arrests = arrest_df.merge(pd_df,left_on='Zip_Codes', right_on='ZIP')
police_arrests.head()
```

```python
police_arrests.count()
```

```python
mhr_df = pd.read_csv('Resources/mhp_mhc_sap.csv')
mhr_df.head()
```

```python
final_data = police_arrests.merge(mhr_df,left_on='Zip_Codes', right_on='zip')
final_data.head()
```

```python
final_data.count()
```

```python
final_data = final_data.drop(columns=['Unnamed: 0','OBJECTID'])
final_data.count()
```

```python
final_data.dtypes
```
