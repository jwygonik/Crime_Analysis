```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
import pandas as pd


# Import our input dataset
ml_df = pd.read_csv('encoded_data.csv')
loans_df.head()
```

```python
ml_cat_1 = ml_df.dtypes[ml_df.dtypes=='object'].index.tolist()
ml_cat_2 = ml_df.dtypes[ml_df.dtypes=='int64'].index.tolist()
ml_cat_3 = ml_df.dtypes[ml_df.dtypes=='float64'].index.tolist()
ml_cat = ml_cat_1 + ml_cat_2 + ml_cat_3
ml_df[ml_cat].nunique()
```

```python
y = ml_df.Sex_Code_M
X = ml_df.drop(columns=['Sex_Code_M'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier. * typically 128 is the highest # of estimators you should use
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)

# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f" Random forest predictive accuracy: {accuracy_score(y_test,y_pred):.3f}")
```
