```python
import pandas as pd 
import datetime as dt 
import numpy as np
import geopy
```

```python
la_df = pd.read_csv('Arrest_Data_from_2020_to_Present.csv',low_memory=False,index_col='Report ID')
la_df.head(10)
```

```python
la_df['Arrest Date'] = pd.to_datetime(la_df['Arrest Date'])
```

```python
la_df_new = la_df.drop(columns=['Location','Report Type','Time','Area ID','Reporting District','Charge Group Code','Arrest Type Code','Address','Charge','Cross Street','Booking Date','Booking Time','Booking Location Code'])
la_df_new.head(10)
```

```python
for col in la_df_new:
  print(la_df_new['Area Name'].unique())
```

```python
for col in la_df_new:
  print(la_df_new['Booking Location'].unique())
```

```python
la_df_new.dropna(inplace=True,axis=0)
la_df_new.count()
```

```python
la_df_new.dtypes
```

```python
la_filtered_df = la_df_new.loc[(la_df_new['Arrest Date']>='01/01/2020')
                           &(la_df_new['Arrest Date']<'01/01/2021')]
la_filtered_df.sort_values('Arrest Date',ascending=True)
```

```python
la_filtered_df.to_csv('la_crimes_2000.csv', index = False)
```
