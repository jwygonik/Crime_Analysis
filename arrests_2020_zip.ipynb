```python
import pandas as pd
import numpy as np
```

```python
arrests_df = pd.read_csv('la_crimes_2000.csv')

lapd_df = pd.read_csv('LAPD_Police_Stations.csv')
```

```python
# df.loc[df[‘column’] condition, ‘new column name’] = ‘value if condition is met’
def zipcodes(x):
    if x == '77th Street':
        return '90003'
    elif x == 'Pacific':
        return '90066'
    elif x == 'Southwest':
        return '90062'
    elif x == 'Hollenbeck':
        return '90033'
    elif x == 'Van Nuys':
        return '91401'
    elif x == 'Central':
        return '90014'
    elif x == 'Devonshire':
        return '91324'
    elif x == 'Newton':
        return '90011'
    elif x == 'Wilshire':
        return '90019'
    elif x == 'Southeast':
        return '90061'
    elif x == 'Hollywood':
        return '90028'
    elif x == 'Harbor':
        return '90731'
    elif x == 'Foothill':
        return '91331'
    elif x == 'Northeast':
        return '90065'
    elif x == 'Rampart':
        return '90017'
    elif x == 'West Valley':
        return '91335'
    elif x == 'Topanga':
        return '91304'
    elif x == 'N Hollywood':
        return '91601'
    elif x == 'Mission':
        return '91345'
    elif x == 'West LA':
        return '90025'
    elif x == 'Olympic':
        return '90006'
    else:
        return 'NaN'
```

```python
arrests_df['Zip_Codes'] = arrests_df['Area Name'].apply(zipcodes)
arrests_df.head(20)
```

```python
arrests_df['Zip_Codes'].isna().sum()
```

```python
arrests_df['Zip_Codes'] = arrests_df['Zip_Codes'].astype(int)
arrests_df['LAT'] = arrests_df['LAT'].astype(float)
arrests_df['LON'] = arrests_df['LON'].astype(float)
```
