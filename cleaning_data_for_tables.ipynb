```python
def pt(x):
    if x == mhp_df['zip'].any()& x == sap_df['zip'].any():
        return 'Both'
    elif x == mhp_df['zip'].any()& x != sap_df['zip'].any():
        return 'General Mental Health Program'
    else:
        return 'Substance Abuse Program'
```

```python
mhc_zip = mhc_df['zip'].to_list()
mhp_zip = mhp_df['zip'].to_list()
sap_zip = sap_df['zip'].to_list()
```
